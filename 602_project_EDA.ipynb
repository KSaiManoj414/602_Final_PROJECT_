# Feature Engineering 


Adding extra column ACCIDENT_DATETIME which will be the combination of ACC_DATE and ACC_TIME
We could use this column in the future to get the overview of Accident happened at what Date,year,month, Time

```python
vehicle_crashes_data['ACCIDENT_DATETIME'] = pd.to_datetime(vehicle_crashes_data.ACC_DATE.astype(str)+' '+vehicle_crashes_data.ACC_TIME.astype(str))
vehicle_crashes_data.head(2)
```

```python
#dropping the columns that have major null values
vehicle_crashes_data.drop(['MUNI_DESC','RTE_SUFFIX','REFERENCE_SUFFIX'],axis='columns',inplace=True)
```

```python
vehicle_crashes_data['COLLISION_TYPE_DESC'].value_counts()
```

By the above most of the crashes happened in the same direction rear end and it is very unlikely for the crashes to happen opposite direction both taking a left turn

```python
#no of accidents w.r.t collision type
sb.set_style('darkgrid', {'grid.color' : '.8', 'grid.linestyle' : ':'})
sb.countplot(x='COLLISION_TYPE_DESC', data = vehicle_crashes_data, palette = 'rainbow')
sb.set(rc = {'figure.figsize':(10,10)})
plt.xticks(
    rotation=45, 
    horizontalalignment='right',
    fontweight='light',
    fontsize='medium'  
)
plt.title('Counting the no. of accidents w.r.t collision type')
```

By the above most of the crashes happened in the same direction rear end and it is very unlikely for the crashes to happen opposite direction both taking a left turn

```python
lights_type = vehicle_crashes_data['LIGHT_DESC'].value_counts()
lights_type
```

```python
#Replacing light_desc with meaning names
#https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.replace.html
vehicle_crashes_data['LIGHT_DESC'] = vehicle_crashes_data['LIGHT_DESC'].replace({'Dawn':'Daylight', 'Dusk':'Dark No Lights', 'Dark - Unknown Lighting':'Dark No Lights'})
```

```python
lights_type = vehicle_crashes_data['LIGHT_DESC'].value_counts()
lights_type
```

```python
lights_type.plot.pie(figsize=(10,10),fontsize=12)
plt.title('CRASHES WITH LIGHT TYPE')
```

From the above pie graph the accidents that happened during day light are far more than the accidents which happened during night with no lights.
